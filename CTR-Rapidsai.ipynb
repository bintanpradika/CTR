```python
file = 'day_15.csv'
header = ['col'+str(i) for i in range (1,41)]
with open(file) as f:
    nrows_file = sum(1 for line in f)
nrows_file
```

### cudf

```python
import cudf
%time dfcudf = cudf.read_csv(file, delimiter='\t', names=header, nrows = nrows_file)
dfcudf
```

```python
dfcudfpandas = dfcudf.to_pandas()
%time type(dfcudfpandas)
```

```python
%time dfcudfpandas
```

### blazingsql

```python
from blazingsql import BlazingContext
%time bc = BlazingContext()
%time bc.create_table('CTR', file, delimiter='\t', names=header, nrows = nrows_file)
%time dfsql = bc.sql('SELECT * FROM CTR')
dfsql
```
